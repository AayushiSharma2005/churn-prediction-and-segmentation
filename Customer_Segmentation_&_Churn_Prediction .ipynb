**Approach 2 combination of trees and depth **

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,recall_score,precision_score,f1_score
```

```python
n_estimators_list=[100,200,300,400,500]
max_depth_list=[5,10,15,20]
results=[]
for n_trees in n_estimators_list:
    for depth in max_depth_list:
        rf_model=RandomForestClassifier(n_estimators=n_trees,
      max_depth=depth,random_state=42,class_weight='balanced')
        rf_model.fit(X_train,Y_train)
        y_pred=rf_model.predict(X_test)
        accuracy=accuracy_score(Y_test,y_pred)
        recall=recall_score(Y_test,y_pred)
        precision=precision_score(Y_test,y_pred)
        f1=f1_score(Y_test,y_pred)
        results.append({'Trees':n_trees,'Depth':depth,'Accuracy':accuracy,'Recall':recall,'Precision':precision,'F1':f1})
results_df=pd.DataFrame(results)
results_df=results_df.sort_values(by=['Recall','Accuracy'],ascending=False)
print(results_df.head(20))
#
```

## **Customer Segmentation**

```python
segmentation_data=pd.DataFrame({'Tenure Months':X['Tenure Months'],
                                'Monthly Charges':X['Monthly Charges'],
                                'Total Charges':X['Total Charges'],
                                'Churn Probability':churn_probability})
```

```python
segmentation_data
```

**Kmeans**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaled_data=scaler.fit_transform(segmentation_data)
```

```python
print(scaled_data[:5])
```

```python
from sklearn.cluster import KMeans
```

```python
wcs=[]
for k in range(1,16):
  kmeans=KMeans(n_clusters=k,random_state=42)
  kmeans.fit(scaled_data)
  wcs.append(kmeans.inertia_)

plt.figure(figsize=(12,6))
plt.plot(range(1,16),wcs, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,random_state=42)
```

```python
clusters=kmeans.fit_predict(scaled_data)
```

```python
segmentation_data['Cluster']=clusters
```

```python
segmentation_data
```

```python
cluster_summary=segmentation_data.groupby('Cluster').mean()
cluster_summary
```

```python
cluster_names={
    0:'Budget Loyal Customers',
    1:'High Risk New Customers',
    2:'Loyal Premium Customers'
}
```

```python
segmentation_data['Cluster Segment']=segmentation_data['Cluster'].map(cluster_names)
```
